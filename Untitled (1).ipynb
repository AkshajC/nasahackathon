```python
traffic_accidents_modified_df = traffic_accidents_df
traffic_accidents_modified_df["latitude"] = round(traffic_accidents_df["latitude"], 1)
traffic_accidents_modified_df["longitude"] = round(traffic_accidents_df["longitude"], 1)
```

```python
traffic_accidents_df["latitude"]
```

```python
traffic_accidents_modified_df.shape
```

```python
traffic_accidents_modified_grouped_df = traffic_accidents_modified_df.groupby(by=["latitude", "longitude"]).agg("count").reset_index()
traffic_accidents_modified_grouped_df.head()
```

```python
# traffic_accidents_modified_grouped_df.get_group(34)

traffic_accidents_agg = traffic_accidents_modified_grouped_df[["latitude", "longitude", "objectid"]]
traffic_accidents_agg.head()
```

```python
traffic_accidents_agg.shape
traffic_accidents_agg["accident_count"] = traffic_accidents_agg["objectid"]
traffic_accidents_agg.drop(columns=("objectid"))
```

```python
traffic_accidents_agg = traffic_accidents_agg.sort_values(by="accident_count", ascending=False)
```

```python
traffic_accidents_agg_high_risk = traffic_accidents_agg.iloc[0:10, :]
traffic_accidents_agg_high_risk
```
